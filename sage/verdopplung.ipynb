```python
# E ist ein Paar [a,b], mit a,b Koeffizienten der Weierstrass-Gleichung
# P ist ein Paar [x,y] mit Koordinaten auf E

def duplicate_point(E,P) :        
    
    # Verdoppelung des Nullpunkts ist der Nullpunkt
    if P == 0 : 
        return 0 
        
    a = E[0]
    b = E[1]
    x = P[0]
    y = P[1]
    
    # Ist die Diskriminante \not=0 ?
    assert(-16*(4*a^3+27*b^2)!=0)
    
    # Ist der Punkte auf der elliptischen Kurve?
    assert(y^2 == x^3+a*x+b)
        
    if (2*y).is_unit() == False :
        return 0        
        
    xx = (9*x^4+6*a*x^2-8*y^2*x+a^2)/(4*y^2)
    yy = (-27*x^6-27*a*x^4+36*y^2*x^3-9*a^2*x^2+12*a*y^2*x-8*y^4-a^3)\
         /(8*y^3)
    
    return [xx,yy]
```
